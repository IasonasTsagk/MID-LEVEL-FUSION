```python
import os
import pandas as pd
from glob import glob

lesion_type_dict = {'akiec': 'Actinic keratoses',
                    'bcc': 'Basal cell carcinoma',
                    'bkl': 'Benign keratosis-like lesions ',
                    'df': 'Dermatofibroma',
                    'nv': 'Melanocytic nevi',
                    'mel': 'Melanoma',
                    'vasc': 'Vascular lesions'}

path1 = ('/Users/iasonastsagkaris/Desktop/archive')
path2 = ('/Users/iasonastsagkaris/Desktop/archive')
path3 = ('HAM10000_metadata.csv')
imageid_path_dict = {}
for path in [path1, path2]:
    for root, dirs, files in os.walk(path):
        for file in files:
            if file.endswith(".jpg"):
                imageid_path_dict[os.path.splitext(file)[0]] = os.path.join(root, file)

skin_df = pd.read_csv(path3)
skin_df['path'] = skin_df['image_id'].map(imageid_path_dict.get)
skin_df['cell_type'] = skin_df['dx'].map(lesion_type_dict.get)
skin_df['hasMelanoma'] = (skin_df['cell_type'] == 'Melanoma').astype(int)
#skin_df['cell_type_idx'] = pd.Categorical(skin_df['cell_type']).codes
skin_df.groupby(['dx']).count()
```
