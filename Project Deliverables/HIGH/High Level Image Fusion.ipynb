```python
# We need to make the two datasets have equal size, so we remove the rows of data_col that we removed from data_bw on the 
# preprocessing stage because the plant was out of bounds
df_col.drop([719, 826, 836, 919, 962, 982], axis=0, inplace=True)
df_col
```

```python
df_col = df_col.reset_index(drop=True)
df_col
```

```python
X_bw = df_bw.iloc[:, 19:]
y_bw = df_bw.iloc[:, :4]
```

```python
X_col = df_col.iloc[:, 14:]
y_col = df_col.iloc[:,:4]
```

```python
y1 = np.round((y_bw["Expert 1"] + y_bw["Expert 2"] 
      + y_bw["Expert 3"] + y_bw["Expert 4"])/4, 0)

y2 = np.round((y_col["Expert 1"] + y_col["Expert 2"] + y_col["Expert 3"] 
               + y_col["Expert 4"])/4, 0)
```

```python
y1 = y1.apply(lambda x: 1 if x >= 3 else 0)
y2 = y2.apply(lambda x: 1 if x >= 3 else 0)
```

```python
# Train - Test - Validate Split
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_bw, y1, test_size=0.15, random_state=42)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_col, y2, test_size=0.15, random_state=42)

X_train1, X_val1, y_train1, y_val1 = train_test_split(X_train1, y_train1, test_size=0.15, random_state=42)
X_train2, X_val2, y_train2, y_val2 = train_test_split(X_train2, y_train2, test_size=0.15, random_state=42)
```

```python
def weighted_voting(model, data):
    
    a = model[0].predict_proba(data[0])
    b = model[1].predict_proba(data[1])
    confidence = []
    for index in range(len(a)):
        proba1 = max(a[index][0], a[index][1])
        proba2 = max(b[index][0], b[index][1])
        confidence.append((proba1, proba2))
    confidence = np.asarray(confidence)
    y_pred1 = model[0].predict(data[0])
    y_pred2 = model[1].predict(data[1])
    weighted_voting = []
    for index in range(len(y_pred1)):
        w1 = confidence[index][0]
        w2 = confidence[index][1]
        l1  = y_pred1[index]
        l2  = y_pred2[index]
        weighted_voting.append(np.round((w1*l1 + w2*l2)/(w1+w2),0))
    weighted_voting = np.asarray(weighted_voting)
    return weighted_voting



# Function to create the confusion matrices
def confusion_matrices(model, test_data):
    confusion_matrices = []
    for clf, (X_test, y_test) in zip(model, test_data):
        y_pred = clf.predict(X_test)
        cm = confusion_matrix(y_test, y_pred)
        confusion_matrices.append(cm)
    return confusion_matrices


def bayesian_fusion_predict(model, datapoints, conf_matrices):
    final_predictions = []
    for n_datapoint in range(datapoints[0].shape[0]):
        predictions = [clf.predict([datapoints[i].iloc[n_datapoint].to_numpy()]) for i, clf in enumerate(model)]
        # we are assuming that the prior probabilities are both 0.5
        p_h0 = 0.5
        p_h1 = 0.5
        for conf_mat, pred in zip(conf_matrices, predictions):
            p_0_h0, p_1_h0 = conf_mat[0] / np.sum(conf_mat[0])
            p_0_h1, p_1_h1 = conf_mat[1] / np.sum(conf_mat[1])
            if pred == 0:
                p_h0_0 = (p_0_h0 * p_h0) / (p_0_h0 * p_h0 + p_0_h1 * p_h1)
                p_h1_0 = 1 - p_h0_0 
                p_h0 = p_h0_0
                p_h1 = p_h1_0
            else:
                p_h0_1 = (p_1_h0 * p_h0) / (p_1_h0 * p_h0 + p_1_h1 * p_h1)
                p_h1_1 = 1 - p_h0_1 
                p_h0 = p_h0_1
                p_h1 = p_h1_1
        final_predictions.append(0 if p_h0 > p_h1 else 1)
    return np.array(final_predictions)

```
